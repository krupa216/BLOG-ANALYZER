```python
# ===================== 🔥 HIGH ACCURACY BLOG ANALYZER =====================

# This line installs several Python libraries required for the project.
# - `transformers==4.36.2`: A specific version of the Hugging Face Transformers library for NLP tasks.
# - `torch`: PyTorch, a deep learning framework.
# - `gradio`: A library to easily build web UIs for machine learning models.
# - `scikit-learn`: A popular machine learning library.
# - `keybert`: A library for keyword extraction based on BERT.
# - `sentence-transformers`: A library for sentence embeddings.
# The `-q` flag ensures a quiet installation.
!pip install -q transformers==4.36.2 torch gradio scikit-learn keybert sentence-transformers

# Imports necessary classes and functions from the installed libraries.
from transformers import pipeline
from keybert import KeyBERT
import gradio as gr

# ===================== 🔹 LOAD MODELS =====================
# Initializes a summarization pipeline using a pre-trained BART-large-CNN model from Hugging Face.
# This model is designed for generating summaries of longer texts.
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
# Initializes a question-answering pipeline using a default pre-trained model.
# This model takes a context and a question to extract an answer from the context.
qa_pipeline = pipeline("question-answering")
# Initializes the KeyBERT model for keyword extraction.
kw_model = KeyBERT()

# ===================== 🔥 KEYWORDS =====================
# Defines a function to extract keywords from a given text.
def extract_keywords(text):
    # Uses the KeyBERT model to extract the top 5 keywords.
    keywords = kw_model.extract_keywords(text, top_n=10)
    # Joins the extracted keywords (first element of each tuple) into a comma-separated string.
    return ", ".join([kw[0] for kw in keywords])

# ===================== 🔥 KEY POINTS (SUMMARY) =====================
# Defines a function to get key points from a text by summarizing it.
def get_key_points(text):
    # Summarizes the first 1000 characters of the text, aiming for a summary length between 40 and 120 characters.
    summary = summarizer(text[:1000], max_length=120, min_length=40)[0]['summary_text']
    # Splits the summary into individual sentences (points).
    points = summary.split(". ")
    # Formats each point with a bullet and joins them with newlines, filtering out very short points.
    return "\n".join([f"• {p}" for p in points if len(p) > 5])

# ===================== 🔥 Q&A =====================
# Defines a function to answer a question based on a given context.
def answer_question(context, question):
    # Uses the question-answering pipeline to find the answer within the first 2000 characters of the context.
    result = qa_pipeline(question=question, context=context[:2000])
    # Returns the extracted answer.
    return result['answer']

# ===================== 🔥 MAIN =====================
# Defines the main analysis function that combines keyword extraction and key point summarization.
def analyze(text):
    # Checks if the input text is too short; if so, returns a warning.
    if len(text) < 25:
        return "⚠️ Enter valid blog"

    # Extracts keywords from the text.
    keywords = extract_keywords(text)
    # Gets key points (summary) from the text.
    key_points = get_key_points(text)

    # Returns a formatted string containing both keywords and key points.
    return f"🔑 Keywords:\n{keywords}\n\n📌 Key Points:\n{key_points}"

# ===================== 🔥 UI =====================
# Sets up the Gradio web interface for the blog analyzer.
with gr.Blocks() as app:
    # Displays a Markdown heading for the application title.
    gr.Markdown("## 🔥 Smart Blog Analyzer")

    # Creates a text input box for the user to paste their blog content.
    blog = gr.Textbox(lines=10, placeholder="Paste blog here...")

    # Creates an 'Analyze' button.
    analyze_btn = gr.Button("Analyze")
    # Creates a text output box to display the analysis results.
    output = gr.Textbox(lines=10)

    # Displays a Markdown heading for the Q&A section.
    gr.Markdown("### Ask Question")
    # Creates a text input box for the user to type a question.
    q = gr.Textbox()
    # Creates an 'Answer' button.
    ans_btn = gr.Button("Answer")
    # Creates a text output box to display the answer.
    ans = gr.Textbox()

    # Links the 'Analyze' button click event to the `analyze` function,
    # taking input from `blog` and displaying output in `output`.
    analyze_btn.click(analyze, inputs=blog, outputs=output)
    # Links the 'Answer' button click event to the `answer_question` function,
    # taking input from `blog` (as context) and `q` (as question), and displaying output in `ans`.
    ans_btn.click(answer_question, inputs=[blog, q], outputs=ans)

# Launches the Gradio application.
# `share=True` generates a public URL, useful for sharing in environments like Google Colab.
app.launch(share=True)
```
